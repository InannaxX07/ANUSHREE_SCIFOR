```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
target = iris.target
data['target'] = target
```

```python
print(data.isnull().sum())  # View missing value counts per column
print(data.info())  # Get overall data information
```

```python
sns.countplot(x='target', data=data)
plt.xlabel('Target Class')
plt.ylabel('Count')
plt.title('Distribution of Iris Flower Species')
plt.show()
```

```python
sns.histplot(x='target', data=data, bins=3, kde=False)  # Adjust bin count for better separation
plt.xlabel('Target Class')
plt.ylabel('Proportion')
plt.title('Proportions of Iris Flower Species')
plt.show()
```

```python
sns.lineplot(x='target', y='sepal length (cm)', hue='target', data=data)
plt.xlabel('Target Class')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Distribution by Target Class')
plt.show()
```

```python
sns.displot(data['sepal length (cm)'])  # Histogram for sepal length
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.title('Distribution of Sepal Length')
plt.show()
sns.boxplot(data['sepal length (cm)'])  # Boxplot for sepal length
plt.xlabel('Sepal Length (cm)')
plt.title('Boxplot of Sepal Length')
plt.show()
```
